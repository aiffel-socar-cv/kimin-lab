```python
import os
import cv2
import matplotlib.pyplot as plt

import sys
# 상위 디렉토리의 파일을 import하기 위해서 사용
sys.path.append("..")
from config import *
```

```python
file_list = os.listdir(RESULTS_DIR)
file_list
```

```python
fig = plt.figure(figsize=(10, 50))

number = 10
start = 1

for crt_id in range(start, start + number):
    idx = crt_id - start
    ax = plt.subplot(number, 3, idx * 3 + 1)
    img = cv2.imread(os.path.join(RESULTS_DIR, f'img_{crt_id:04}.png'))
    plt.imshow(img)
    plt.title("Input image")

    ax = plt.subplot(number,3, idx * 3 + 2)
    mask = cv2.imread(os.path.join(RESULTS_DIR, f'label_{crt_id:04}.png'))
    plt.imshow(mask, cmap='gray')
    plt.title("label image")

    ax = plt.subplot(number,3, idx * 3 + 3)
    output = cv2.imread(os.path.join(RESULTS_DIR, f'output_{crt_id:04}.png'))
    plt.imshow(output, cmap='gray')
    plt.title("output image")

```
